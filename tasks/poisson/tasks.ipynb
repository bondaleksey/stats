# Task 6 
Рассмотрим пуассоновский процесс с интенсивностью 2 события в минуту.

Найдите вероятность:
- `P1` того, что за первую минуту произойдет не более одного события, а за первые две минуты произойдет ровно два события.

## решение

Интересующее нас событие может произойти двумя способами:

1. нуль событий в течение первой минуты и два события в течение второй. Вероятность этого, вследствие независимости количества событий на непересекающихся интервалах, равна 
$ e^{-2} \frac{2^2}{2} \cdot e^{-2} = 2e^{-4}$;
2. одно событие в течение каждой из двух минут. Вероятность этого, вследствие независимости, равна $4e^{-4}$

Получаем по теореме сложения 
$ P = 6e^{-4} \approx  0.10989.$
